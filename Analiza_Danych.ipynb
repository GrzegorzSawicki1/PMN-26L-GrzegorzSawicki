**Temat:** Analiza wpływu mediów społecznościowych na zdrowie psychiczne młodzieży z wykorzystaniem metod AI

## 1. Wybór zestawu danych

* **Nazwa zbioru danych:** Social Media Impact on Teen Mental Health
* **Źródło danych:** Platforma Kaggle (autor: `algozee/teenager-menthal-healy`)
* **Liczba rekordów:** 1200
* **Adekwatność do problemu biznesowego/społecznego:** Zbiór danych zawiera szczegółowe informacje na temat zachowań cyfrowych, nawyków związanych ze stylem życia oraz wskaźników psychologicznych nastolatków. Problem badawczy dotyczy identyfikacji wzorców behawioralnych (np. czasu przed ekranem, higieny snu), które negatywnie wpływają na dobrostan psychiczny młodzieży.
* **Zastosowanie metod AI:** Dane posiadają zdefiniowaną i zbalansowaną zmienną celu (`depression_label`) oraz zestaw cech numerycznych i kategorycznych, co pozwala na zastosowanie zaawansowanych algorytmów uczenia maszynowego (klasyfikacja nadzorowana) do wczesnego wykrywania ryzyka depresji.

## 2. Analiza zbioru danych

### Opis i typ cech w zbiorze:

| Nazwa kolumny | Opis cechy | Typ zmiennej |
| :--- | :--- | :--- |
| `age` | Wiek badanego nastolatka | Numeryczna (Dyskretna) |
| `gender` | Płeć (*male*, *female*) | Kategoryczna (Binarna) |
| `daily_social_media_hours` | Dzienny czas spędzony w mediach społecznościowych [h] | Numeryczna (Ciągła) |
| `platform_usage` | Najczęściej używana platforma (*Instagram*, *TikTok*, *Both/Other*) | Kategoryczna (Nominalna) |

| `physical_activity` | Czas poświęcony na aktywność fizyczną dziennie [h] | Numeryczna (Ciągła) |

| `stress_level` | Samodzielnie oceniany poziom stresu (skala 1–10) | Numeryczna (Dyskretna) |
| `anxiety_level` | Poziom lęku i niepokoju (skala 1–10) | Numeryczna (Dyskretna) |
| `addiction_level` | Poziom uzależnienia od smartfona (skala 1–10) | Numeryczna (Dyskretna) |
| `depression_label` | **Zmienna celu:** Występowanie stanów depresyjnych (0 - brak, 1 - obecne) | Kategoryczna (Binarna) |

### Braki w danych:
Wstępna weryfikacja techniczna (`df.info()`) wykazuje, że zbiór jest kompletny. Brak jest wartości pustych (*NaN* / *Missing Values*), co eliminuje konieczność stosowania zaawansowanych technik imputacji (uzupełniania) danych i pozwala na bezpośrednie przejście do analizy korelacji.

### Wstępna analiza korelacji i zależności:
* **Silna korelacja dodatnia:** Długi czas spędzany w social mediach (`daily_social_media_hours`) oraz korzystanie z telefonu bezpośrednio przed snem (`screen_time_before_sleep`) silnie korelują ze wzrostem wskaźników `stress_level`, `anxiety_level` oraz końcowym `depression_label`.
* **Silna korelacja ujemna:** Liczba godzin snu (`sleep_hours`) oraz regularna aktywność fizyczna (`physical_activity`) wykazują działanie protekcyjne – wyższe wartości tych cech współwystępują z niższym poziomem stanów lękowych i depresyjnych.

### Planowane wykresy do sekcji EDA (Exploratory Data Analysis):
1. **Macierz korelacji (Heatmap):** Do zweryfikowania liniowych współczynników korelacji Pearsona między wszystkimi zmiennymi numerycznymi.
2. **Wykresy pudełkowe (Boxplots):** Porównanie dystrybucji poziomu lęku (`anxiety_level`) w zależności od płci oraz preferowanej platformy.
3. **Wykresy gęstości (KDE Plot):** Rozkład czasu snu w grupach zdiagnozowanych jako zagrożone depresją vs grupa zdrowa.

## 3. Sformułowanie hipotezy badawczej

> **Hipoteza:** "Na podstawie profilu behawioralnego nastolatka – uwzględniającego jego nawyki cyfrowe (czas przed ekranem, czas w mediach społecznościowych) oraz styl życia (długość snu, aktywność fizyczna) – modele uczenia maszynowego są w stanie skutecznie przewidzieć i sklasyfikować podwyższone ryzyko wystąpienia stanów depresyjnych (`depression_label = 1`)."

* **Cel modelu:** Klasyfikacja binarna.
* **Zastosowanie praktyczne:** Narzędzie wspierające psychologów szkolnych i pedagogów w procesie wczesnego przesiewu i identyfikacji jednostek wymagających wsparcia psychologicznego na podstawie ankiet zachowań.

## 4. Aktualny stan wiedzy (State of the Art)

Problem wpływu transformacji cyfrowej na zdrowie psychiczne generacji Z oraz Alpha jest obecnie jednym z kluczowych tematów w obszarze *Health Analytics* i psychologii behawioralnej.

1.  **Podejście klasyczne (Statystyczne):** Tradycyjne badania medyczne opierają się głównie na statystyce opisowej, testach ANOVA oraz klasycznej regresji logistycznej. Badania (np. prof. Jean Twenge) udowodniły statystyczny związek między gwałtownym wzrostem posiadania smartfonów przez młodzież a korelatami pogorszenia zdrowia psychicznego. Narzędzia te są jednak ograniczone w wykrywaniu nieliniowych, wieloczynnikowych interakcji.
2.  **Podejście AI / Machine Learning:** Współczesne systemy predykcyjne odchodzą od prostych modeli liniowych na rzecz algorytmów drzewiastych (takich jak *Random Forest*, *XGBoost*, *Gradient Boosting*) oraz sieci neuronowych. Pozwalają one na wykrycie złożonych zależności (np. sytuacji, w której wysoki poziom interakcji społecznych w świecie rzeczywistym całkowicie niweluje negatywny wpływ 3 godzin przed ekranem, ale traci swoje właściwości obronne przy 6 godzinach). W literaturze naukowej kładzie się nacisk na optymalizację metryki **Recall (Czułość)** oraz **F1-Score**, ponieważ w systemach medycznych/diagnostycznych kluczowe jest zminimalizowanie błędów typu II (nieprzeoczenie nastolatka w kryzysie).
