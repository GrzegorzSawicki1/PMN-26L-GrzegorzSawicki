## 4. Aktualny stan wiedzy
1. Wpływ mediów społecznościowych na zdrowie psychiczne młodzieży

Źródło: Keles, B., McCrae, N., & Grealish, A. (2020). A systematic review: the influence of social media on depression, anxiety and psychological distress in adolescents.
* Treść: Badanie to stanowi obszerny przegląd systematyczny literatury dotyczącej wpływu mediów społecznościowych na młodzież. Wyniki wskazują na silną, proporcjonalną korelację między czasem spędzanym na platformach społecznościowych a nasileniem objawów depresji, lęku oraz ogólnego stresu psychologicznego. Zwrócono uwagę na mechanizmy takie jak FOMO (Fear of Missing Out), cyberprzemoc oraz negatywne porównywanie się z wyidealizowanym obrazem rówieśników.
Proponowane podejścia i rozwiązania:

 - ograniczanie ekspozycji na media społecznościowe u nastolatków poprzez kontrolę rodzicielską lub edukację z zakresu cyfrowej higieny,

 - promowanie jakościowych, a nie tylko ilościowych, interakcji w sieci (np. aktywne wsparcie rówieśnicze zamiast biernego scrollowania),

implementowanie przez same platformy algorytmów ostrzegających o nadmiernym czasie użytkowania.
Dlaczego pasuje do tematu: Artykuł bezpośrednio uzasadnia uwzględnienie zmiennych daily_social_media_hours oraz platform_usage w naszym zbiorze danych. Pokazuje, że intensywność korzystania z social mediów jest jednym z głównych zewnętrznych stymulatorów problemów psychicznych u nastolatków. W kontekście naszego modelu pozwala to na sformułowanie założenia, że wysokie wartości w kolumnie daily_social_media_hours będą silnie korelować ze wzrostem parametrów anxiety_level oraz stress_level, ostatecznie wpływając na zmienną docelową depression_label.

2. Teoria wyparcia snu przez urządzenia ekranowe

Źródło: Woods, H. C., & Scott, H. (2016). #Sleepyteens: Social media use in adolescence is associated with poor sleep quality, anxiety, depression and low self-esteem.
* Treść: Badanie analizuje, w jaki sposób korzystanie z urządzeń ekranowych przed snem wypiera czas przeznaczony na odpoczynek (tzw. hipoteza wyparcia – displacement theory). Wykazano, że nocne zaangażowanie emocjonalne w media społecznościowe oraz ekspozycja na światło niebieskie drastycznie obniżają jakość i długość snu, co z kolei stanowi bezpośredni fizjologiczny katalizator dla rozwoju zaburzeń depresyjnych i lękowych.
Proponowane podejścia i rozwiązania:

 - wyznaczanie „stref wolnych od ekranu” (np. sypialni) na minimum godzinę przed planowanym zaśnięciem, psychoedukacja nastolatków na temat krytycznej roli snu w regulacji emocjonalnej i procesach poznawczych, uwzględnianie nawyków związanych ze snem w rutynowych badaniach przesiewowych zdrowia psychicznego młodzieży.

Dlaczego pasuje do tematu: Publikacja ta stanowi kluczowy łącznik (mediator) dla naszego modelu predykcyjnego, nadając ogromne znaczenie zmiennym sleep_hours oraz screen_time_before_sleep. Teoria wyparcia snu pozwala nam oczekiwać, że relacja w danych nie jest jednowymiarowa: wysoki czas przed ekranem (screen_time_before_sleep) redukuje długość snu (sleep_hours), a ten fizjologiczny deficyt bezpośrednio winduje ryzyko klasyfikacji badanej osoby jako depression_label = 1.
